```python
# load data
td = pd.read_csv("BABA.csv")
td.head(10)
```

```python
#print total no of rows in datset
print("Total Number Of Rows :",len(td))
```

```python
# Data Preprosessing and Statistical Summary of our data
# Lets Calculate Some Measurements Of Every Coloumn
#Mean
#S.D
#Minimum Or Maximum Value
print("Total No Of Dates in our dataset :", len(td))

                                     #Open
    #In finance, SD is used to measure the risk associated with an investment.
    #A higher SD indicates higher volatility and thus higher risk.
print("Mean Value Of Our Opening Price Of Stock :", td['Open'].mean())
print("S.D Value Of Our Opening Price Of Stock :", td['Open'].std())
print("Maximum Value Of Our Opening Price Of Stock :", td['Open'].max())
print("Minimum Value Of Our Opening Price Of Stock :", td['Open'].min())
                                     #High
print("Mean Value Of Stock How Much It Go High :", td['High'].mean())
print("SD Value Of Our Coloum High:", td['High'].std())
print("Maximum Highest Value Of Our  Stock :", td['High'].max())
print("Minimum Highest Value Of our Stock :", td['High'].min())
                                     #Low
print("Mean Value Of Stock How Much It Go Low :", td['Low'].mean())
print("SD Value Of Our Coloum Low:", td['Low'].std())
print("Maximum Lowest Value Of Our  Stock :", td['Low'].max())
print("Minimum Lowest Value Of our Stock :", td['Low'].min())
                                    #Close
print("Mean Value Of closing Price Of Stock :", td['Close'].mean())
print("SD Value Of Our Coloum Close:", td['Close'].std())
print("Maximum Closest Value Of Our  Stock :", td['Close'].max())
print("Minimum Closest Value Of our Stock :", td['Close'].min())
                                  #Adj Close 
        #I Will Drop This Coloum Because I dont Need To Predict How Companey Will Used Or split Its Dividend
        #IN A Coorporation
td = td.drop(columns=['Adj Close'])
td.head(10)




```

```python
#Checking Missing Or Null Values
td.isnull().sum()
```

```python
#Convert the 'Date' column into datetime format
td['Date'] = pd.to_datetime(td['Date'])

# Extract year, month, and day as new features
td['Year'] = td['Date'].dt.year
td['Month'] = td['Date'].dt.month
td['Day'] = td['Date'].dt.day

# Drop the original 'Date' column and 'Adj Close' column
td = td.drop(columns=['Date'])

#Split the data into target(y) and feature Variable
Feature_var=['Year','Month','Day','Open','High','Low','Volume']
X=td[Feature_var]
y=td['Close']
td.head(10)
```

```python
#split the data into four variables x_train,y_train ,x_test,Y-test
#Testing Size 20%
#Training Size 80%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
# Create and train the model
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
# It Means Value Is Not Predicted Accurately
 #Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
#Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute Error:', mae)
mae = mean_absolute_error(y_test, y_pred)
#Calculate R1 Score
r2 = r2_score(y_test, y_pred)
r3=r2*100
print(f'R-squared: {r3}')
```

```python
# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
#Visualizes the actual vs predicted values with a scatter plot.
```
